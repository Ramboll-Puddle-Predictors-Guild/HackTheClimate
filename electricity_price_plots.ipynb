```python
import polars as pl

data = pl.read_csv("european_wholesale_electricity_price_data_hourly/Denmark.csv").lazy()


data = data.select(pl.col("*").exclude(["Datetime (UTC)", "Datetime (Local)", "Country", "ISO3 Code"]), time=pl.col("Datetime (Local)").str.to_datetime("%Y-%m-%d %H:%M:%S", time_unit="ms"))
data = data.filter(pl.col("time").dt.year() < 2020).drop_nulls()
data = data.set_sorted("time")
# data = data.group_by_dynamic(
#     pl.col("time"),
#     every="1",
#     period="1mo"
# ).agg(pl.col("*").exclude("time").mean())

data = data.collect()

data.plot("time", "Price (EUR/MWhe)")
```
